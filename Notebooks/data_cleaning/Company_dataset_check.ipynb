```python
import pandas as pd
import numpy as np
import os
```

```python
path_5min = "E:/Karan_Bais/python/Ml/Projects/stock/Datasets/Company_datasets/5min"
path_3min = "E:/Karan_Bais/python/Ml/Projects/stock/Datasets/Company_datasets/3min"
path_15min = "E:/Karan_Bais/python/Ml/Projects/stock/Datasets/Company_datasets/15min"
dataset_names = [ "HDFCBANK", "ICICIBANK", "INFY", "RELIANCE", "SBIN", "TCS"]
```

## read all datasets

```python
datasets_5min = {}
datasets_15min = {}
datasets_3min = {}
for name in dataset_names:
    datasets_5min[name] = pd.read_csv(os.path.join(path_5min, (name+'.csv')))
    datasets_3min[name] = pd.read_csv(os.path.join(path_3min, (name+'.csv')))
    datasets_15min[name] = pd.read_csv(os.path.join(path_15min, (name+'.csv')))
```

# check for nulls

## 5min

```python
for name,dataset in datasets_5min.items():
    print(name,'\n',dataset.isna().sum(), '\n========================================================================')
```

## 3min

```python
for name,dataset in datasets_3min.items():
    print(name,'\n',dataset.isna().sum(), '\n========================================================================')
```

## 15min

```python
for name,dataset in datasets_15min.items():
    print(name,'\n',dataset.isna().sum(), '\n========================================================================')
```
